Mel Frequency Cepstral Coefficents (MFCCs) is a way of extracting features from an audio. The MFCC uses the MEL scale to divide the frequency band to sub-bands and then extracts the Cepstral Coefficents using Discrete Cosine Transform (DCT).


The MFCC is a good algorithm to extract high frequency and low frequency information.

### Audio Framing

An audio frame or sample contains amplitude (loudness) information at that particular point in time. To produce sound, tens of thousands of frames are played sequentially to produce frequencies.

Audio is a non-stationary process, so the FFT will produce distortions.
So we can assume the audio is a stationary process for a short periods of time.

We have to segment the audio into short frames. Each segment will be the same size as the FFT and also with a frame overlap.
The overlap on the frames will have some correlation between them, reason been the applciation of a window function(Hann window) which makes information to be lost on the edges of each frame.

```python
def frame_audio(audio, n_fft=h.n_fft, hop_size=10, 
                  sample_rate=None):
    
    # hop_size is in ms
    audio = np.pad(audio, int(n_fft / 2), mode='reflect')
    
    frame_len = np.round(sample_rate * hop_size / 100).astype(int) # 1600
    
    frame_num = int((len(audio) - n_fft) / frame_len) + 1  # 19
    
    frames = np.zeros((frame_num, n_fft))
    
    for i in range(frame_num):
        frames[i] = audio[i*frame_len:i*frame_len+n_fft] 
    
    
    return frames
```

```python
hop_size = 15 # ms

audio_framed = frame_audio(audio, h.n_fft, hop_size, sample_rate)
print(f'Framed audio shape: {audio_framed.shape} ')
```

Now, we have a framed audio matrix with the size of frame number X n_fft size

### Convert to frequency domain

Convert audio from time domain to frequency domain. FFT assumes audio to be periodic an continues. By framing the signal we assured the audio to be periodic.

To make it continues we apply a window function on every frame to break the high frequency distortions. The window assures that both ends of the signal will end close to zero.

NB: We need to apply the window on the framed audio before we can perform FFT.

```python
# using the Hanning Window
window = get_window('hann', h.n_fft, fftbins=True)

plt.figure(figsize=(15, 4))
plt.plot(window)
plt.grid(True)
```

```python
audio_win = audio_framed * window

index = 4
plt.figure(figsize=(15, 6))
plt.subplot(2, 1, 1)
plt.plot(audio_framed[index])
plt.title('Original Frame')
plt.grid(True)
plt.subplot(2, 1, 2)
plt.plot(audio_win[index])
plt.title('Frame After Windowing')
plt.grid(True)
```

 You can see that the window brought the edges of each frame closer to zero.
 
 Now perorm the FFT. The we only take the positive part of the spectrum(first half).

```python
audio_winT = audio_win.T

# reason for the np.complex64 is that, there might be a situation of sqrt(-1),
# which cant be dealt with so we use an imaginary symbol for that.
audio_fft = np.empty((int(1 + h.n_fft // 2), audio_winT.shape[1]),
                    dtype=np.complex64, order='F')

for i in range(audio_fft.shape[1]):
    audio_fft[:, i] = fft.fft(audio_winT[:, i], axis=0)[:audio_fft.shape[0]]
    
audio_fft = audio_fft.T
```

### Calculate signal Power

```python
audio_power = np.square(np.abs(audio_fft))
print(audio_power.shape)
```

### MEL-spaced filterbank

Compute the Mel-space filter banks and pass the framed audio through them.
This will give us information about the power in each frequency band.

Mel filter banks helps to get the different frequency from the normal fft and convert them the normal frequency in hertz. MEL-spaced filterbanks is the spacing between the filters grows exponentially with frequency.

```python
f_min = 0 # frequency minimum
f_max = sample_rate / 2  # frequency maximum
mel_filter_num = 10

print(f'Minimum freq.: {f_min}')
print(f'Maximum freq.: {f_max}')
```

#### Compute filter points

1. first convert the two filterbank edges to the MEL space. After that we construct a lineary spaced array between the two MEL frequencies. 

2. Then we convert the array to the frequency space and finally we normalize the array to the FFT size and choose the associated FFT values.

```python
def freq_to_mel(freq):
    return 2595.0 * np.log10(1.0 + freq / 700.0)

# inverse
def mel_to_freq(mels):
    return 700.0 * (10.0**(mels / 2595.0) - 1.0)
```

```python
def get_filter_points(fmin, fmax, mel_filter_num,
                     fft_size, sample_rate=sample_rate):
    fmin_mel = freq_to_mel(fmin)
    fmax_mel = freq_to_mel(fmax)
    
    print(f'MEL min: {fmin_mel}')
    print(f'MEL Max: {fmax_mel}')
    
    mels = np.linspace(fmin_mel, fmax_mel, num=mel_filter_num+2)
    freqs = mel_to_freq(mels)
    
    
    return np.floor((fft_size + 1) / sample_rate * freqs).astype(int), freqs
    
    
```

```python
filter_points, mel_freqs = get_filter_points(f_min, f_max,  mel_filter_num,
                                             h.n_fft, sample_rate)
filter_points
```

#### Construct the fiterbank

```python
def get_filters(filter_points, fft_size):
    
    filters = np.zeros((len(filter_points)-2, int(fft_size/2+1)))
    
    for i in range(len(filter_points)-2):
        filters[i, filter_points[i] : filter_points [i + 1]] = \
                    np.linspace(0, 1, filter_points[i + 1] - filter_points[i])
        filters[i, filter_points[i + 1]: filter_points [i + 2]] = \
                    np.linspace(1, 0, filter_points[i + 2] - filter_points[i + 1])
        
    return filters
```

```python
filters = get_filters(filter_points, h.n_fft)

plt.figure(figsize=(15, 4))

for i in range(filters.shape[0]):
    plt.plot(filters[i])
plt.ylabel('Weight')
plt.xlabel('Freq.')
plt.title('Mel freq.')
```

Next divide the triangular MEL weights by the width of the MEL band(area normalization).
If the filters are not normalized, the noise will increase with frequency because of the filter width.

Weight is like a filter sound. If weight is equal to 1 , then dont touch the signal. When freq is high, we spread the distance to match the original spectrogram distance.
Lower end frequency means, the frequency is muted.


```python
# taken from librosa lib.

enorm = 2.0 / (mel_freqs[2:mel_filter_num+2] - mel_freqs[:mel_filter_num])
filters *= enorm[:, np.newaxis]  # mel filter shape, its in 2d
```

```python
plt.figure(figsize=(15, 4))
for i in range(filters.shape[0]):
    plt.plot(filters[i])


plt.ylabel('Weight')
plt.xlabel('Freq.')
plt.title('Normalized Mel freq.')
```

#### Filter the signal

```python
audio_filtered = np.dot(filters, audio_power.T)

audio_log = 10.0 * np.log10(audio_filtered) # since we perceive audios in decibels

audio_log.shape
```

### Generate the cepstral Coefficiencts

Scale for pitch. 
We generate the MFCC by using the Discrete Cosine Transform (DCT). It'll extract the high frequency and low frequency changes in the signal.

```python
def dct(dct_filter_num, filter_len):
    basis = np.empty((dct_filter_num, filter_len))
    basis[0, :] = 1.0  / np.sqrt(filter_len)
    
    samples = np.arange(1, 2 * filter_len, 2) * np.pi / (2.0 * filter_len)
    
    for i in range(1, dct_filter_num):
        basis[i, :] = np.cos(i * samples) * np.sqrt(2.0 / filter_len)
        
    return basis
```

```python
dct_filter_num = 40

dct_filters = dct(dct_filter_num, mel_filter_num)

cepstral_coefficients = np.dot(dct_filters, audio_log)

cepstral_coefficients.shape
```


Why MFCC?
1. Simplified version of fourier transform.
2. Gets real-valued coefficient
3. Decorrelate energy in different mel bands or banks.
    - since theres overlapping among each frequency, theres
    correlation.
4. Reduce number of dimensions to represent a spectrum.

Note: The higher the index of MFCC the higher the signal we try to fix into the log spectrum.
